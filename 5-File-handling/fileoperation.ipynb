### File operation - Read adn write file

Fil handling is a crucial part of any programming language. Python provide build in functions and methods to read and write to files, both text and binary. This lesson will cover the basic of file handling, include reading and writing text file and binary file.

```python
### Read A whole file 

with open('example.txt', 'r') as file:

    content = file.read()
    print(content)
```

```python
## Read a file line by line

with open('example.txt', 'r') as file:
    for line in file:
        # print(line)
        print(line.strip())  # Remove leading/trailing whitespace
```

```python
## Writing a file(overwriting)

with open('example.txt', 'w') as file:
    file.write("Hello work \n")
    file.write("This is a new line \n")
```

```python
## Write a file without overwriting 

with open('example.txt', 'a') as file:
    file.write("This line will be added to the end of the file \n")
```

```python
### Writing a list of line to a file

lines = ["First line \n", "Second line \n", "Third line \n"]

with open('example.txt', 'a') as file:
    file.writelines(lines)
```

```python
## Binary files 

# Write to a binary file
data = b'\x00\x01\x02\x03\x04'
with open('example.bin', 'wb') as file:
    file.write(data)
```

```python
# Read from a binary file

with open('example.bin', 'rb') as file:
    content = file.read()
    print(content)
```

```python
## Read a content from source text file and write to a destination text file
# Copying a text file
with open('example.txt', 'r') as source_file:
    content = source_file.read()

with open('destination.txt', 'w') as destintion_file:
    destintion_file.write(content)
```

```python
# Read a text file andcount the number of lines, words and characters.
# Counting lines, words, and characters in a atext file

def count_text_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()
        num_lines = len(lines)
        num_words = sum(len(line.split()) for line in lines)
        num_characters = sum(len(line) for line in lines)

    return num_lines, num_words, num_characters

file_path = 'example.txt'
lines, words, characters = count_text_file(file_path)
print(f"Lines: {lines}", f"Words: {words}", f"Characters: {characters}")
```

The W+ mode in python is used to opena a file for both reading and writing. If the file does not exist, it will be created. If the file exists, its countent is truncated(i.e., The file is overwritten)

```python
### Writing and then reading a file

with open('example.txt', 'w+') as file:
    file.write("Hello word\n")
    file.write("This is a new line\n")

    ## Move the file cursor to the begning 
    file.seek(0)

    ## Read the content of the file
    content = file.read()
    print(content)
```

```python
## Checking if a path is a file or directory
import os 
path = 'example.txt'
if os.path.isfile(path):
    print(f"The path '{path}' is a file.")
elif os.path.isdir(path):
    print(f"The path '{path}' is a directory.")
else:
    print(f"The path '{path}' is neither a file nor a directory.")
```
