## Import module in python: Modules and Packages

In Python, modules and packages help organize and resuse code. Here's comprehensive guide on how to import them.


```python
import math 
math.sqrt(16)
```

```python
from math import sqrt, pi
print(sqrt(16))
print(sqrt(25))
print(pi)
```

```python
import numpy as np
np.array([1, 2, 3])
```

```python
from math import *
print(sqrt(16))
print(pi)
```

```python
from package.math import addition
addition(2, 3)
```

```python
from package import math
math.addition(2, 3)
```
