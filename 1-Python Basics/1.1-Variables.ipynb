### Variable
Variable are fundamental element in programming used to store data that can be referd and manipulated in a program.In Python, variable are created when you assign a vlue to them, they do not need explicit declaration to reserve memory space. the declaation happenns automatically when you assign a value a variable.


```python
## Declaring and assigning variable 

age = 32
height = 6.1
name = "Akshansh"
is_student = True

## Printing a variable 

print("age: ", age)
print("Height: ", height)
print("Name: ", name)
```

```python
## Naming convention
## Variable name should be descriptive
## They must start with a letter or an '_' and contains letter, number and underscore
## Variable name case sensitive

## Valid name 
first_name = "Akshansh"
last_name = "Dogra"

## Invalid name

#2age = 32
#first-name = "Akshansh"
#@name = "Akshansh"

##Case sensitive
name = "Akshansh"
Name = "Akshansh"
name == Name # False
```

```python
## Understanding variable types 
## pyhon is dynamically typed, type of variable is determined at runtime

age = 25 #int 
height= 6.1 #float 
name = "Akshansh" #string
is_student = True #boolean

print(type(age))
print(type(name))
```

```python
## Type Checking and Type Conversion

age  = 25
print(type(age))

# Type conversion
age_str = str(age)
print(age_str)
print(type(age_str))
```
