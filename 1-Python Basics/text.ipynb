### Indentation 
Indentation in Python is used to define the structure and hierarchy of the code. Unlike many other programming laguages that use braces{} to delimit block of code, Python uses indentation to determine the grouping of statement, This means that all the statements within a block must be indented at the same level.

```python
## Indentation 
## Python uses indentation to define block of code. consistent use of spaces (commonly 4) or a tab is required. space after the colon or if statement for pint line.

age  = 32 
if age>30:
    print(age)

print(age)
```
