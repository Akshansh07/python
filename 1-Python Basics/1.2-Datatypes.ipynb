### DataType

Data types are classification of data which tell the compiler or interpreter how the programmer intendes to use the data 

They determine the type of operattions that can be perform on the data, the values that the data can take, and the amount of memory needed to store the data.


```python
## Integer Example 

age  = 25
type(age)

##floating point 
width  = 4.5
type(width)

## Sting datatype
name = "Akshansh"
type(name)


## Boolean datatype
is_student = True
type(is_student)
```
