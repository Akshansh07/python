## Condition statement 


```python
## elif
## the elif statement allows you to check multiple conditions. It stands for "else if"

age = 17

if age<13:
    print("You are a child.")
elif age < 18:
    print("You are a teenager.")
else: 
    print("You are an adult.")
```

```python
## Nested Condition Statements
# You can place one or more if, elif, or else statement inside another if, elif, or else statement to create nested conditional statements.

## Number even, odd, negative

num = int(input("Enter a Number: "))

if num > 0: 
    print("The number is positive")
    if num%2 == 0: 
        print("The number is even")
    else: 
        print("The number is odd")

else:
    print("The number is negative") 
```

```python
## Pratical Example 

## Detemine if the year is a Leap year using nested contion statement

year = int(input("Enter The Year:"))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print(year, "Is a leap year")
        else:
            print(year, "Is not a leap year")

    else: 
        print(year, "is a leap year")

else:
    print(year, "is not a leap year")
```
