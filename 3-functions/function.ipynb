### Functions 

A function is a block if code that perform a specific task. functions help in organizing code, resusing code, and improve redability.


```python
## Syntax

def functions_name(parameter):
    """Docstring"""
    # Function body
    return expression
```

```python
## Why functions?

num = 24
if num%2 == 0:
    print("the number is even")
else:
    print("the number is odd")
    
```

```python
def even_or_odd(num):
    """This function perform finds even and odd"""
    if num%2 == 0:
        print("the number is even")
    else:
        print("number is false")
```

```python
## Call this function 

even_or_odd(24)
```

```python
## Function with multiple parameter 

def add(a, b):
    return a+b

result = add(2,4)
print(result)
```

```python
## Default Parameters

def greet(name):
    print(f"Hello {name} Welcome to the paradise")

greet("Akshansh")
```

```python
## Default Parameters

def greet(name = "Guest"):
    print(f"Hello {name} Welcome to the paradise")

greet("aks")
```

```python
### Variable length argument 
## Positional and keyword argument

def print_number(*args):
    for number in args:
        print(number)
```

```python
print_number(1,2,3,4,5,6,7,8,"akshansh")
```

```python
## Positional argument

def print_number(*args):
    for number in args:
        print(number)
```

```python
print_number(1,2,3,4,5,6,7,8,"akshansh")
```

```python
### Keyword argument

def print_details(**kwargs):
    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(name="akshansh", age="35", country="india")
```

```python
def print_details(*args, **kwargs):
    for val in args:
        print(f" Positional argument :{val}")

    for key,value in kwargs.items():
        print(f"{key}:{value}")
```

```python
print_details(1,2,3,name="akshansh", age="35", country="india")
```
