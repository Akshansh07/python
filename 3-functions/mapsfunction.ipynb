### The map() Function in python 

The map() function applies a given function to all items in an inout list(or any other iterable) and return a map object(an iterator). This is particularly useful for transforming data in a list comprehensively

```python
def square(x):
    return x*x

square(2)
```

```python

numbers = [1,2,3,4,5,6,7,8]

list(map(square,numbers))
```

```python
## Lambda function with map

list(map(lambda x:x*x, numbers))
```

```python
### Can we map multiple iterable

number1 = [1,2,3]
number2 = [4,5,6]

added_numbers = list(map(lambda x,y: x+y, number1, number2))
print(added_numbers)
```

```python
##  map to convert list of string to integer

str_numbers = ["1","2","3","4","5"]
int_numbers = list(map(int, str_numbers))
print(int_numbers)
```

```python
Words = ["apple", "banana", "cherry"]
upper_word = list(map(str.upper, Words))

print(upper_word)
```

```python
def get_name(person):
    return person['name']

people = [
    {'name': 'Alice', 'age': 30},
     {'name': 'Aksahnsh', 'age': 25},
]

list(map(get_name, people))
```
