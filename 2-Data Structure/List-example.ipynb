### Manage a to do list 



```python
to_do_list = ["Buy groceries", "Clean the house", "Pay bills"]

## Add to task

to_do_list.append("Schedule Meeting")
to_do_list.append("Go for the run")

## Remove completed task
to_do_list.remove("Clean the house")

## Checking if the task in the list
if "Pay bills" in to_do_list:
    print("Don't forget to pay the bills!")

print("Todo list remaning")
for task in to_do_list:
    print(f"-{task}")
```

Organizing student grage

```python
# Oarganizing student grade 

grades = [85, 92, 78, 90, 88]

# Adding new grade
grades.append(95)

# Calculate average grade
average_grade = sum(grades) / len(grades)
print(f"Average Grade: {average_grade:.2f}")

# Find highest and lowest grade 

heighest_grade = max(grades)
lowest_grade = min(grades)
print(f"Higest Grade: {heighest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```

Managing an inventery

```python
# Manage an inventery
inventory = ["apple", "Banana", "orange", "grapes"]

# Adding a new item

inventory.append("watermelon")

# Removing an item
inventory.remove("Banana")

# check if item in a stock 

item = "Orange"

if item in inventory:
    print(f"{item} is in stock.")
else:
    print(f"{item} is out of stock.")


# Print the inventory 

print("Inventery List:")
for item in inventory:
    print(f"- {item}")
```

Collect User Feedback 


```python
# Collect user feedback

feedback =  ["Great service!", "Very Satisfied", "Could be better", "Excellent experience"]

# Adding new feedback
feedback.append("Not happy with the product")

# Count specific feedback
positive_feedbak_count = sum(1 for comment in feedback if "great" in comment.lower() or "excellent" in comment.lower())
print(f"Positive FeedackCount: {positive_feedbak_count}")

# Print all feedback 

for comment in feedback:
    print(f" - {comment}")
```
