### Dictionsries

Dictionaries are unordered collections of items. They store data in key value pairs. Keys must be unquie and immutable, While value can be of any type.

```python
## creating Dictonaries 

empty_dict = {}
print(type(empty_dict) )
```

```python
empty_dict = dict()
empty_dict
```

```python
student = {"name": "Akshansh", "age": 35, "grade": 24}
print(student)
```

```python
# Error 
# Single key is always used.

student = {"name": "Akshansh", "age": 35, "name": 24}
print(student)
```

```python
## Accessing dictionary values

student = {"name": "Akshansh", "age": 35, "grade": "A"}
print(student)
```

```python
print(student["grade"])
print(student["age"])

## Accesing ussing get() method

print(student.get("grade"))
print(student.get("last_name"))
print(student.get("last_name", "Not Found"))
```

```python
## Modification of dictionary values
## Dictionaries are mutable, so you can add, update and delete elements.

print(student)
```

```python
student["age"] = 33   ## update vale
print(student)
student["addess"]="India"  ## add new key-value pair
print(student)
```

```python
del student["grade"] ## delete key-value pair
print(student)
```

```python
## Dictionary Methods

keys = student.keys()  ##get all keys of the dictionary
print(keys)
values = student.values() ##get all values of the dictionary
print(values)
items = student.items() ## get all key-value pairs of the dictionary
print(items)
```

```python
## Shallow copy

student_copy = student
print(student)
print(student_copy)
```

```python
student["name"] = "akshansh2"
print(student)
print(student_copy)
```

```python
student_copy1 = student.copy() ## create a shallow copy of the dictionary
print(student_copy1)
print(student)
```

```python
student["name"] = "Rahul"
print(student)
print(student_copy1)
```

```python
## Iterating in dictonaries
## you can use loops to iterates over dictionaries, keys, value or items.

## Iterating over keys 

for keys in student.keys():
    print(keys)
```

```python
## Iterate over values 

for value in student.values():
    print(value)
```

```python
## Iterate over key value pair 

for key, value in student.items():
    print(f"{key}: {value}")
```

```python
## nested dictonaries

students = {
    "student1": {"name": "akshansh", "age": 25},
    "student2": {"name": "rahul", "age": 30}
}
print(students)
```

```python
## Access nested dictionary element 
print(students["student1"] ["name"])
print(students["student1"] ["age"])
```

```python
## Interating over nested dictionary

for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key, value in student_info.items():
        print(f"{key}: {value}")
```

```python
## Dictionary comprehension

squares = {x: x**2 for x in range(5)}
print(squares)
```

```python
## Conditional dictionary comprehension 

even = {x: x**2 for x in range(10) if x %2 == 0}
print(even)
```

```python
## Practical example.

## Use a dictionary to count the frequency of element in list

numbers = [1,2,2,2,3,3,4,4,4,4]
frequency = {}

for number in numbers:
    if number in frequency:
        frequency[number] += 1
    else:
        frequency[number] = 1

print(frequency)
```

```python
## Merge 2 dictionaries into one 

dict1 = {"a": 1, "b": 2}
dict2 = {"b": 3, "c": 4}

merged_dict = {**dict1, **dict2}
print(merged_dict)
```
