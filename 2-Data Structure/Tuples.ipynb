### Tuples

Tuples are ordered collection of iteams that are immutable. They are similar to list, but their immutability makes them different.

```python
## Creating a tuple

empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
lst  = list()
print(type(lst))
tpl = tuple()
print(type(tpl))
```

```python
numbers = tuple([1, 2, 3, 4, 5, 6])
numbers
```

```python
list((1, 2, 3, 4, 5, 6))
```

```python
mixed_tuple = (1, "Hello Word", 3.14, True)
print(mixed_tuple)
```

```python
## Accessing tuple items 

numbers
```

```python
## Same slicing and indexing operation same as list [0:4], [::-1].
```

```python
## Tuple operation 

concatenation_tuple = numbers + mixed_tuple
print(concatenation_tuple)
```

```python
mixed_tuple * 3
```

```python
numbers * 3
```

```python
## Immutable Nature of Tuple
## Tuples are immutable, meaning their elements cannot be changed once assign.

lst = [1, 2, 3, 4, 5, 6]
print(lst)

lst[1] = "Akshansh"
print(lst)
```

```python
numbers[1] = "Akshansh"
```

```python
## Tuple Methods 

print(numbers.count(1))
print(numbers.index(3))
```

```python
## Paking and unpacking of tuple

packed_tuple = 1, "hello", 3.14
print(packed_tuple)
```

```python
## Unpacking a tuple

a, b, c = packed_tuple
print(a)
print(b)
print(c)
```

```python
## Unpacking with star *

numbers = (1, 2, 3, 4, 5, 6)
first,*middle, last = numbers
print(first)
print(middle)
print(last)
```

```python
## Nested Tuples 


## Nested tuple 
lst = [[1, 2,3,4],[5,6,7,8], [1, "Hellow", 3.14, "c"]]
lst[0][3]
lst[0][0:3]
```

```python

lst = [[1, 2,3,4],[5,6,7,8], (1, "Hellow", 3.14, "c")]
lst[2][0:3]
```

```python
nested_tuple = [(1, 2,3,4), ("a", "b", "c"), (True, False)]

## Access the element inside the tuple 
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
## Iterating over nested tuple

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item, end=" ")
    print()
```

### Conclusions 

Tuples are versatile and useful in many real-word scenarios where an immutable and ordered collection of items are required. they are commonly used in data structure, functions arguments and return values, and as dictionary keys. Understanding how to leverge tuples effectively can improve the efficiency and readability of ypur python code.
