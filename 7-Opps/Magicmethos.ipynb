### Magic Methods 

Magic methods in python, also known as dunder methods(double underscore methods), are special methods that start and end with double underscores. These methods enable you to define the behaviour of objects for built-in operations, such as arithmetic operations, comparison, and more.

### Magic merthods 

 Magic merthods are predefined methods in python that you can override to change of behaviour of your objects. Some common magic methods include:
 

```python
'''

__init__ : Initialize a new instance of the class.
__str__ : Return a string representation of the object.
__repr__ : Return an official string representation of the object.
__len__ : Return the length of the object.
__getitem__ : Gets an item from a contaier.
__setitem__ : Sets an item in a container.

'''
```

```python
## Basic magic methods

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = Person("John", 30)
print(person)
```

```python
## Basic magic methods

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


    def __str__(self):
        return f"Person(name={self.name}, age={self.age})"

    def __repr__(self):
        return f"Person(name={self.name}, age={self.age})"

person = Person("John", 30)
print(person)
print(repr(person))
```
