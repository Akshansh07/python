### Encapsulation and Abstraction 

Encapsulation and abstraction are two fundamental principal of Object Oriented Programming(oop) that help in designing robust, maintainable and reusable code. Encapsulation involves bundling data and methods that operate in the data within a single unit, while abstraction involves hiding complex implementation details and exposing only the necessary feature.

### Encapsulation 

Encapsulation is the concept of wrapping data (variable) and methods (function) together as a single unit. It restricts direct access to some of the object components, which is a means of preventing accidental interference and misuse of the data.

```python
### Encapsulation with Getter and Setter Methods
### Public, protected and private variables or access modifiers.

class Person:
    def __init__(self, name, age):
        self.name = name  # Public variable
        self._age = age   # Protected variable

def get_name(person):
    return person.name

person = Person("Alice", 30)
print(person.name)  # Accessing public variable
get_name(person)  # Accessing public variable through getter function
```

```python
dir(person)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self.__name = name  # Private variable
        self.__age = age     # Private variable
        self.gender = gender

def get_name(person):
    return person.__name

person = Person("Akshansh", 25, "Male")
get_name(person)  # Accessing private variable through getter function
```

```python
person = Person("Akshansh", 25, "Male")
dir(person)
```

```python
class Person:
    def __init__(self, name, age, gender):
        self._name = name  # Protected variable
        self._age = age     # Protected variable
        self.gender = gender

class Employee(Person):
    def __init__(self, name, age, gender):
        super().__init__(name, age, gender)



employee = Employee("Akshansh", 25, "Male")
print(employee._name)  # Accessing protected variable from subclass
```

```python
## Encapsulation with Getter and Setter Methods
class Person:
    def __init__(self, name, age):
        self.__name = name # Private variable
        self.__age = age     # Private variable

    ## Getter method for name 
    def get_name(self):
        return self.__name
    
    ## Setter method for name
    def set_name(self, name):
        self.__name = name

    ## Getter method for age
    def get_age(self):
        return self.__age

    ## Setter method for age
    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("Age cannot be negative.")

person = Person("akshansh", 25)

### Accessing and modify variable using getter and setter 

print(person.get_name())
print(person.get_age())

person.set_age(30)
print(person.get_age())

person.set_age(-5)  # Invalid age, should print error message
```
