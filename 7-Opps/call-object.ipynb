### Classes and oObjectd

Object-oriented programming(oop) is a programming paradigm that uses "object"  to design application and computer program. OOP allows for modeling real-word scenarios using classes and objects. This lesson covers the basic of creating classes and objects, include instance variable and method as.

```python
### A class is a blue print for creating objects. Attributes, methods 

class Car:
    pass

audi = Car()
bmw = Car()

print(type(audi))
```

```python
### Instance variable and method

class Dog:
    ## Constructor
    def __init__(self, name, age):
        self.name = name
        self.age = age

## Create object 

dog1 = Dog("Buddy", 3)
print(dog1)
print(dog1.name)
print(dog1.age)


```

```python
dog2 = Dog("Charlie", 5)
print(dog2.name)
print(dog2.age)
```

```python
## Define class with instance method

class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} says Woof!")

dog1 = Dog("Buddy", 3)
dog1.bark()
dog2 = Dog("Lucy", 8)
dog2.bark()
```

```python
### Modeling a Bank Acoount 

## Define a class for bank account
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner    ### Instance variable
        self.balance = balance

        ## This init part is a constructor

    def deposit(self, amount):
        self.balance += amount ### Instance method
        print(f"{amount} is deposited. New balance: {self.balance}")
    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds")
        else:
            self.balance -= amount
            print(f"{amount} is withdrawn. New balance: {self.balance}")

    def get_balance(self):
        return self.balance
    
## Create an object
account = BankAccount("Akshansh", 500)
print(account.balance)
```

```python
## Call instance method
account.deposit(200)
```

```python
account.withdraw(200)
```

```python
print(account.get_balance())
```
