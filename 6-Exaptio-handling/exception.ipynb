### Understanding Exaption 

Exception handling in python allows you to handle errors gracefully an take corrective action without stopping the execution of the program. This lesson will cover the basics of excaption, include how use try, except and finally blocks.


### What Are Exceptions?

Exception are events that disrupt the normal flow of a program. They occur when error is encountered during program execution. Common exeption error.
1. ZeroDivisionError: Divided by zero
2. FileNotFoundError: File not found
3. ValueError: Invalid value.
4. TypeError: Invalid type, 

```python
## Exception try, except block

try: 
    a = b
except:
    print("Variable not been assigned")
```

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Please enter a valid number")
except ZeroDivisionError:
    print("Please enter a number greater than zero")
except Exception as ex:
    print(ex)
    print("An unexpected error occurred")
```

```python
### try, except, else block

try: 
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Please enter a valid number")
except ZeroDivisionError:
    print("Please enter a number greater than zero")
except Exception as ex:
    print(ex)
else: 
    print(f"The result is {result}")
```

```python
## try, except, else and finally block

try: 
    num = int(input("Enter a number: "))
    result = 10 / num
except ValueError:
    print("Please enter a valid number")
except ZeroDivisionError:
    print("Please enter a number greater than zero")
except Exception as ex:
    print(ex)
else: 
    print(f"The result is {result}")
finally:
    print("Execution completed.")
```
